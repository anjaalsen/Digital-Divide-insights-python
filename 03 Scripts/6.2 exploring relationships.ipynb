```python
# check for missing values again
df_internet.isnull().sum()
```

```python
# check for many missing data in rows
df_internet.isnull().sum(axis=1).value_counts()
```

```python
# display rows with 4 missing data
# Set pandas option to display all rows
pd.set_option('display.max_rows', None)

# Filter rows with 4 missing values and display the 'Country Name' column
rows_with_4_missing = df_internet[df_internet.isnull().sum(axis=1) == 4]
print(rows_with_4_missing['Country Name'])
```

```python
# Remove rows with more than 3 missing values; 
# the affected countries make up 13.7% of the entire datset and are distributed globally and economically, 
# ensuring their exclusion will not introduce significant bias into the analysis.

df_internet = df_internet[df_internet.isnull().sum(axis=1) <= 3]
```

```python
# identify numerical columns in the dataset

numerical_cols = df_internet.select_dtypes(include=['float64', 'int64']).columns
```

```python
# Filter rows with up to 3 missing values
df_filtered = df_internet[df_internet.isnull().sum(axis=1) <= 3]

# Check numerical columns with missing values
missing_numerical_cols = numerical_cols[df_filtered[numerical_cols].isnull().any()]
print(missing_numerical_cols)
```

```python
# Impute missing values for identified numerical columns

for col in missing_numerical_cols:
    df_internet.loc[df_internet.isnull().sum(axis=1) <= 3, col] = df_internet.loc[df_internet.isnull().sum(axis=1) <= 3, col].fillna(df_internet[col].mean())
```

## 3. Exploring Relationships
### Correlations

```python
# Create a correlation matrix using pandas
#Select only the numerical columns from the DataFrame to avoid errors
numerical_columns = df_internet.select_dtypes(include=['int', 'float']).drop(columns=['Year'])

numerical_columns.corr()
```

```python
# Create a correlation heatmap using matplotlib

plt.matshow(numerical_columns.corr())
plt.show()
```

```python
# Add labels, a legend, and change the size of the heatmap

f = plt.figure(figsize=(9, 9)) # figure size 
plt.matshow(numerical_columns.corr(), fignum=f.number) # type of plot
plt.xticks(range(numerical_columns.shape[1]), numerical_columns.columns, fontsize=9, rotation = 90) # x axis labels
plt.yticks(range(numerical_columns.shape[1]), numerical_columns.columns, fontsize=9) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14) # add font size
plt.title('Correlation Matrix', fontsize=14) # add title
```

```python
# Create a categorical plot in seaborn using the user categories created above

# Set the style
sns.set(style="ticks")

# Define a custom color palette with green and orange
custom_palette = {"Lower Internet Use": "green", "Higher Internet Use": "orange"}

# Create the categorical plot
g = sns.catplot(
    x="Region", 
    y="Internet_Users", 
    hue="User category", 
    data=df_internet, 
    palette=custom_palette,  # Apply the custom color palette
    kind="strip",  # Default type, you can change to 'swarm', 'violin', etc.
    height=7,  # Adjust the figure size
    aspect=1.5  # Adjust the aspect ratio
)

# Add labels and title
g.set_axis_labels("Region", "Internet Users (% of Population)")
g.fig.suptitle("Internet Users by Region and User Category", y=1.02, fontsize=14)

# Format the y-axis ticks explicitly
max_y = df_internet["Internet_Users"].max()
plt.yticks(np.arange(0, max_y + 10, step=10))  # Set ticks every 10 units

plt.xticks(rotation=45)

# Remove unnecessary elements
g.despine(left=True)

# Save the plot as a PNG file
output_path = r"C:\Users\User\250104_Global Internet Access Analysis\04 Analysis\catplot.png"
g.savefig(output_path, dpi=300, bbox_inches='tight')
```

### Interpretation of categorical plot:
### Regions like Europe & Central Asia and North America have predominantly higher internet usage, while regions like Sub-Saharan Africa have a significant concentration of lower internet usage, indicating disparities in global internet access.

## 4. Answering the first research questions, others need further exploration

#### - How does GDP per capita influence the percentage of individuals using the internet?
The correlation heatmap shows a positive correlation (0.73) between GDP per capita and internet users, indicating that wealthier countries tend to have higher internet adoption rates. This suggests that economic prosperity supports the development of internet infrastructure and affordability.

#### - What is the impact of literacy rates (total, male, and female) on internet adoption?
The heatmap reveals moderate correlations between total literacy (0.41), male literacy (0.29), and female literacy (0.44) with internet usage. This suggests that literacy is an important factor for internet adoption, but other variables (e.g., infrastructure) likely play a stronger role.

#### - Which regions or income groups are the furthest behind in internet adoption, and what common factors contribute to this gap?
The categorical plot by region indicates that Sub-Saharan Africa and South Asia have the highest concentration of lower internet use. These regions often have lower GDP, limited infrastructure, and lower electricity access, contributing to the gap.

## 5. Defining hypotheses

#### Hypothesis 1:
Countries with higher GDP per capita will have significantly higher internet usage than countries with lower GDP per capita.

#### Hypothesis 2:
Regions with higher broadband subscriptions and electricity access will exhibit higher internet adoption than regions with lower infrastructure levels.

#### Hypothesis 3:
Countries with both high literacy rates and strong infrastructure will have higher internet adoption compared to countries with only one of these factors.

#### Hypothesis 4:
Sub-Saharan Africa and South Asia have the largest gaps in internet adoption due to a combination of economic and infrastructure constraints.

#### Hypothesis 5:
There will be a stronger correlation between internet adoption and life expectancy in regions with higher GDP compared to regions with lower GDP.
