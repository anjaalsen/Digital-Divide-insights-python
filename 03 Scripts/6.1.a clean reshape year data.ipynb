## 2. Wrangle data

```python
# Filter data to exclude years 1960-2011
# Define the columns to keep: metadata and years 2012 to 2022
year_columns = [str(year) for year in range(2012, 2023)]
keep_columns = ['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'] + year_columns

# Filter columns for each dataset in the list
filtered_datasets = []
for df in dataset_list:
    filtered_df = df[keep_columns]  # Keep only the required columns
    filtered_datasets.append(filtered_df)
```

```python
print(filtered_datasets[0].head())  
print(filtered_datasets[0].shape)
```

```python
# Reshape from wide to long format
# Reshape each dataset in filtered_datasets to long format
long_datasets = []

for df in filtered_datasets:
    # Use pd.melt() to reshape the dataframe
    long_df = df.melt(
        id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'],  # Columns to keep as-is
        var_name='Year',  # Name of the new column for the years
        value_name='Value'  # Name of the new column for the corresponding values
    )
    long_datasets.append(long_df)  # Append the reshaped dataframe to the list
```

```python
long_datasets[1].columns
```

```python
# Define the common set of columns
#common_columns = ['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code', 'Year', 'Value']

# Ensure all DataFrames have the same columns
#for i in range(len(long_datasets)):
#    long_datasets[i] = long_datasets[i][common_columns]

# Combine all long datasets into a single dataframe
combined_long_dataset = pd.concat(long_datasets, ignore_index=True)
```

```python
value_counts = combined_long_dataset['Indicator Name'].value_counts()
print(value_counts)
```

```python
# Pivot the dataset to have one column per indicator
final_dataset = combined_long_dataset.pivot(
    index=['Country Name', 'Country Code', 'Year'],  # Columns to keep as index
    columns='Indicator Name',  # Column to pivot into multiple columns
    values='Value'  # Values to fill the new columns
)

# Flatten the column index (after pivot, columns become a MultiIndex if multiple levels exist)
final_dataset.columns = [str(col) for col in final_dataset.columns]  # Flatten columns
final_dataset = final_dataset.reset_index()  # Reset the index to turn it into regular columns

print(final_dataset.shape)  # Check the shape of the dataset
print(final_dataset.head(50))  # Inspect the first few rows
```
